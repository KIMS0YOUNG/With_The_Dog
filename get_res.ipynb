# 식당 정보를 얻어오는 코드
#### CSV 파일을 DataFrame 형식으로 가져와서 처리 후 다시 CSV 파일로 저장


```python
import pandas as pd
# 데이터 프레임 형식으로 CSV 파일 열기
# "서울 식당 기본 정보" 파일
res_basic = pd.read_csv("/content/drive/MyDrive/WithTheDog/data/서울관광재단_식당기본정보_20211222.csv")
# "서울 식당 운영 정보" 파일: 반려동물 동반 가능 여부 확인 가능
res_manage = pd.read_csv("/content/drive/MyDrive/WithTheDog/data/서울관광재단_식당운영정보_20211231.csv")

# 내용 확인. 각각 처음부터 5개, 마지막부터 5개
print(res_basic)
print(res_manage)
```

```python
# 열 인덱스 확인
basic_col_idx = res_basic.columns
manage_col_idx = res_manage.columns
print("###기본정보###\n", basic_col_idx)
print("###운영정보###\n", manage_col_idx)
```

```python
# 필요없는 열 삭제
res_manages_drop = res_manage.drop( ['와이파이제공여부', '놀이방유무', '다국어메뉴판제공여부', '화장실정보내용',  '택배판매유무', '배리어프리유무',
       '배달서비스유무', '예약방식명', '온라인예약정보내용', '홈페이지(URL)', '인근랜드마크명', '인근랜드마크위도',
       '인근랜드마크경도', '인근랜드마크와거리', '키오스크유무', '모바일페이먼트유무', '스마트오더유무', '입식좌석수',
       '좌식좌석수', '대표메뉴명'], axis= "columns")
print(res_manages_drop) 

# "기본정보" 데이터는 식당/카페/주요 판매 항목을 알아보기 위해 가져옴
# 그러나 "운영정보" 데이터와 겹치는 부분이 많고, 메뉴(ID)로 정보를 가져오기 어려워 사용하지 않기로 함.
```

```python
# 특정 요소를 가진 행만 출력
condition = res_manages_drop.반려동물입장가능여부 == "Y" # 조건 정하기
res_with_dog=res_manages_drop[condition] # 조건에 맞는 행만 출력(반려동물입장가능한 식당만 남기기)

print(res_with_dog)
```
